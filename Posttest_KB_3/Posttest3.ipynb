### Visualisasi Data

Langkah Pertama yang dilakukan adalah mengimport library yang akan digunakan :

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Setelah di import

Read untuk CSV yang akan digunakan dan lihat value dari columns agar dapat digunakan dalam chart :

```python
df = pd.read_csv("msft.csv")
df['Category'].value_counts()
```

### Grafik Tipe Bar

visualisasi data yang pertama saya gunakan memakai Bar yang mana akan di masukkan value dari columns 
untuk penentuan x nya, kemudian pada y akan diberikan jumlah masing masing pemilihan kateogri tersebut.
untuk font disini saya custom agar tidak terlalu monoton.

```python
category = ['Music','Books','Business','Health and Fitness','Developer Tools '
            ,'Lifestyle','News and Weather','Kids and Family','Social','Food and Dining'
            ,'Navigation and Maps','Multimedia Design','Government and Politics']
font = {'family':'serif','color':'blue','size':15}
plt.figure(figsize=(30,8),dpi=80)
plt.bar(category,df['Category'].value_counts())
plt.title('Grafik Batang Kategori Windows Store',size =20,fontdict = font)
plt.xlabel('Category',size=14,fontdict = font)
plt.ylabel('Jumlah Pemilihan Kategori',size=14,fontdict = font)
plt.show()
```

### Grafik Tipe Pie

visualisasi data ke-2 yang saya pakai adalah pie chart, dengan menggunakan autopct untuk memperlihatkan presentasi dari jumlah yang telah ada.
kemudian saya mengggunakan label distance untuk mengatur jaraknya serta rotatelabel untuk membuat nya terbalik.

```python
label = 'Music','Books','Business','Health and Fitness','Developer Tools ','Lifestyle','News and Weather','Kids and Family','Social','Food and Dining','Navigation and Maps','Multimedia Design','Government and Politics'
font = {'family':'serif','color':'blue','size':15}
fig, axs = plt.subplots(figsize=(20,15))
axs.pie(df['Category'].value_counts(),radius=2,  labeldistance=0.64, labels=label, 
              rotatelabels =True, startangle=180,counterclock=False, autopct='%0.1f%%')
axs.axis('equal')
axs.legend(category, title='List Kateogry', loc=0)
axs.set_title('Pie Chart Kategori Aplikasi',size=16,fontdict = font)
plt.xlabel('Presentasi Kategori yang dipilih',size=16,fontdict = font)
plt.xticks(rotation = 45)
plt.show()
```

### Grafik Tipe Line

untuk grafik ke-3 saya  menggunakan garis. yaitu dengan membuat jumlah pemilihan menjadi garis nya atau nilai x nya.dan list kategori untuk nilai y nya

```python
font = {'family':'serif','color':'blue','size':15}
plt.figure(figsize=(20,10))
plt.plot(df['Category'].value_counts(), category, label ='Garis')
plt.legend()
plt.title('Grafik Line Kategori Windows Store',size =20,fontdict = font)
plt.xlabel('Jumlah Pemilihan Kategori',size=14,fontdict = font)
plt.ylabel('List Kategori',size=14,fontdict = font)
plt.legend()
plt.show()
```

### Grafik Tipe Stem

Untuk grafik ke-4 sini sebenarnya hanya percobaan,karena saya mengambil diluar dari modul. pertama disini saya bikin axis nya kemudian langsung memanggil fungsi stemnya setelah itu dimasukkan
nilai nilai masing parameternya menggunakan variabel yang telah diatur. sama seperti grafik garis, untuk nilai x nya menggunakan jumlah pemilihan dan untuk nilai y nya menggunakan list kategori

```python
font = {'family':'serif','color':'blue','size':15}
fig, ax = plt.subplots(figsize=(16,14))
ax.stem(df['Category'].value_counts(), category,linefmt='grey', markerfmt='*')
plt.xlabel('Jumlah Pemilihan Kategori',size=14,fontdict = font)
plt.title('Grafik Batang Kategori Windows Store',size =20,fontdict = font)
plt.ylabel('List Kategori',size=14,fontdict = font)
plt.show()
```
